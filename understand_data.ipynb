```python
%matplotlib notebook
import pandas as pd
import numpy as np
```

```python
fn = './data/297K.sdf'
fn_d = './data/onlydata.sdf'
```

```python
with open(fn_d, 'w') as f_out:
    with open(fn, 'r') as f:
        lines = f.readlines()
        for l in lines:
            if (not '=' in l) and ('\t' in l):
                f_out.write(l)
```

```python
df = pd.read_csv(fn_d, sep='\t', names=['d1', 'd2'])
```
